```python
import pandas as pd
import os
```

```python
def sanitize_repo_name(repo_name):
    """Sanitize repository name by replacing '/' and '-' with '_'"""
    return repo_name.replace('/', '_').replace('-', '_')
```

```python
ck_results_dir = "./ck_results"
summary_results_path = "summarized_results.csv"

df_summary = pd.read_csv(summary_results_path)

loc_data = []

# Percorre os arquivos de CK para obter LOC
for repo_name in df_summary['name']:
    class_csv_path = os.path.join(ck_results_dir, f"{sanitize_repo_name(repo_name)}class.csv")

    if os.path.exists(class_csv_path):
        df_ck = pd.read_csv(class_csv_path)

        loc_total = df_ck['loc'].sum()

        loc_data.append({'name': repo_name, 'loc_total': loc_total})
    else:
        print(f"⚠️ CK output missing for {repo_name}, skipping LOC analysis.")
        loc_data.append({'name': repo_name, 'loc_total': None})

df_loc = pd.DataFrame(loc_data)

df_final = df_summary.merge(df_loc, on="name", how="left")

print("✅ LOC added to summarized_results.csv successfully!")
```
